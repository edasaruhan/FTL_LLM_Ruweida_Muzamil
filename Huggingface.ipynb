```python
from transformers import pipeline
```

```python
# Load the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
#texts about sustainable agriculture
texts = [
    "Sustainable farming practices are essential for protecting our environment.",
    "Using fewer chemicals in agriculture leads to healthier crops and soil.",
    "There are challenges in adopting sustainable agriculture, but the benefits outweigh the difficulties.",
    "Some farmers are resistant to change and prefer traditional methods over sustainable ones."
]

```

```python
results = sentiment_analyzer(texts)


for text, result in zip(texts, results):
    print(f"Text: {text} => Sentiment: {result['label']} with score {result['score']:.4f}")
```
